```python
# Rename columns df2
df2.rename(columns={
    'WHO Region': 'who_region',
    'ISO3': 'iso3',
    'WHO Country Name': 'country_name',
    'City or Locality': 'city',
    'Measurement Year': 'year'
}, inplace=True)
```

```python
df2.head()
```

```python
df1['city'] = df1['city'].str.strip()#That code is cleaning up the 'city' column in both df1 and df2 by removing leading and trailing
df2['city'] = df2['city'].str.strip()
```

```python
df1['year'] = df1['year'].fillna(-1).astype(int)#That code is handling missing values (NaN) in the 'year' column of both df1 and df2.

df2['year'] = df2['year'].fillna(-1).astype(int)
```

```python
df1 = df1.dropna(subset=['year'])# "This code removes missing values in the 'year' column from both DataFrames and then converts the 'year' values to integers."
df2 = df2.dropna(subset=['year'])
df1['year'] = df1['year'].astype(int)
df2['year'] = df2['year'].astype(int)
```

```python
df1['year'] = df1['year'].astype(int)
#astype(int) is used to convert the values in the 'year' column to integers.
df2['year'] = df2['year'].astype(int)
```
